```python
from gtts import gTTS
from googletrans import Translator
from IPython.display import Audio

# 1. Original English text
text_en = """Python Functions is a block of statements that does a specific task.
The idea is to put some commonly or repeatedly done task together and make a function,
so that instead of writing the same code again and again for different inputs
we can do the function calls to reuse code contained in it over and over again.
We can define a function in Python, using the def keyword."""

# 2. Translate
translator = Translator()
text_mr = translator.translate(text_en, src='en', dest='hi').text

# 3. Convert to speech
tts = gTTS(text=text_mr, lang='hi', tld='com')
tts.save("speech.wav")

# 4. Play
Audio("speech.wav", autoplay=False)
```

```python
import pyttsx3
from IPython.display import Audio

text = '''After creating a function in Python we can call it by using the name of the functions 
Python followed by parenthesis containing parameters of that particular function. '''

audio = pyttsx3.init()
audio.setProperty('rate', 150)
audio.setProperty('volume', 0.8)

# change the voices
voice = audio.getProperty('voices')

# 0 for male and 1 for female
#audio.setProperty('voice', voice[0].id)      # for male voice   
audio.setProperty('voice', voice[1].id)       # for female voice

# text-to speech conversion
audio.say(text)

# save the audio file
#audio.save_to_file(text, 'test_male_Voice.mp3')
audio.save_to_file(text, 'test_female_Voice.mp3')

audio.runAndWait()
```
